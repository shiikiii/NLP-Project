For Occasion: <br/>
We consider following columns to identify occasions: product_full_name, description, brand_category, details.<br/>
There are five categories under occasions:
* Casual / Weekend
* Day to Night
* Night Out
* Work
* Workout

```python
# Fill Nan with /, which may block the string operation
target_columns = ['product_full_name', 'description', 'brand_category', 'details']
for col in target_columns:
    full_data[col] = full_data[col].fillna('/')
```

```python
# join full_data and tagged_data to have some sense
join_data = full_data.merge(tagged_data, on='product_id')
join_data[join_data['attribute_name']=='occasion'][target_columns+['attribute_value']].head(5)
```

```python
join_data['tag_info'] = join_data[target_columns+['brand']].apply(lambda x: ' '.join(x), axis=1)
```
